```python
plt.style.use('ggplot')

ax = hints_stats_df['count'].value_counts().plot(kind='barh', figsize=(18,12),
                                                 color="slateblue", fontsize=13);
ax.set_alpha(0.8)
ax.set_title("no.of times notification popup", fontsize=18)
ax.set_xlabel("frequency_count", fontsize=18);
ax.set_xticks([0, 5])

# set individual bar lables using above list
for i in ax.patches:
    # get_width pulls left or right; get_y pushes up or down
    ax.text(i.get_width()+.1, i.get_y()+.31, \
            str(round((i.get_width()), 2)), fontsize=15, color='dimgrey')

# invert for largest on top 
ax.invert_yaxis()
```
