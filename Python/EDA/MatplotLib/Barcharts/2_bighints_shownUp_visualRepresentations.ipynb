```python
weekday_df =hour_df.loc[~hour_df['week_day'].isin(['Sat','Sun'])]
#reset_index(inplace=True)
weekday_df = weekday_df[['_id','hour']]
weekday_hour_count_df = weekday_df.groupby('hour').size().reset_index(name='count')
weekday_hour_count_df.head()
```

```python
weekday_hour_count_df['count'].sum()
```

```python
from matplotlib import pyplot as plt
ax = plt.gca()
weekday_hour_count_df.plot(kind='bar',x='hour',y='count',color='yellowgreen',ax=ax)
plt.rcParams["figure.figsize"] = (15,8)
plt.show()
```
