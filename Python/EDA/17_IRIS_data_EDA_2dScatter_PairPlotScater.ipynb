```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
iris_path = './iris_data.csv'
```

```python
iris_df = pd.read_csv(iris_path)
iris_df.shape
```

```python
iris_df.columns
```

```python
iris_df.info()
iris_df.dtypes
```

```python
#iris_df.info()
iris_df['variety'].value_counts()  
```

```python
setosa_df = iris_df[iris_df['variety']=='Setosa']
virginica_df = iris_df[iris_df['variety']=='Virginica']
versicolor_df = iris_df[iris_df['variety']=='Versicolor']

setosa_df.head()
virginica_df.head()
versicolor_df.head()
```

```python
setosa_df.plot(kind='scatter',x='sepal.length',y='sepal.width');
plt.show()
 
```

```python
virginica_df.plot(kind='scatter',x='sepal.length',y='sepal.width',grid=True);
plt.show()
```

```python
versicolor_df.plot(kind='scatter',x='sepal.length',y='sepal.width',grid=True);
plt.show()
```

```python
iris_df.plot(kind='scatter',x='sepal.length',y='sepal.width',grid=True);
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(iris_df,hue="variety",height=4)\
    .map(plt.scatter, "sepal.length","sepal.width")\
    .add_legend();
plt.figure(figsize=(30,18))
plt.show()
```

# Pair - Plot 


```python
#Pair wise scatter plot : pair -plot
#Dis-advanatages :
#can be used when no.of features are high.
#cannot visualize higher dimensional patterns in 3-D , 4-D.
#Only possible to view 2D Patterns.

plt.close()
sns.set_style('whitegrid')
sns.pairplot(iris_df,hue='variety',height=3);
plt.show()
```
