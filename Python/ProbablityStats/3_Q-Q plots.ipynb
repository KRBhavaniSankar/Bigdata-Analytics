1. Q-Q plot : If X and Y are random variables and with Q-Q plot we can able to figure out following 
    (Q) : is X follows N(mean(meau),variance (sigma square))
    (Q) : If X and Y are random variables then does X and Y have the same distribution ?
    

```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
std_normal = np.random.normal(loc=0,scale=1,size=1000)
std_normal
```

```python
for i in range(1,101):
    print(i,np.percentile(std_normal,i))
    
```

```python
measurements = np.random.normal(loc=20,scale=5,size=50000)
stats.probplot(measurements,dist="norm",plot=pylab)
pylab.show()
```

# Observations

1. As far is increase my samples size more and more points are closer to the line.
2. So if sample size is sample Q-Q plot will not work and it is hard to interpret Q-Q plot. This is the major limitation of Q-Q plot.
3. Here Both X and Y axis follows Normal distribution, that is the reason why both are at same points and increases accordignly.  So you can see constant change in both X and Y.

```python
# Generate  100 samples from Uniform distribution
measurements = np.random.uniform(low=-1,high=1,size=10000)
stats.probplot(measurements,dist="norm",plot=pylab)
pylab.show()
```

1. Here X follows gaussian distribution and Y follows uniform distribution.
2. As far as X increases and we are seeing significant deviation on both (x,y) positions.  So more are more points deviating from straight line.
3. So if both points are at same position and on a straight line then we can conclude both x and y are from same family.
